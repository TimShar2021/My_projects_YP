## Подготовка данных и выбор модели

### Подготовка данных

Удалим лишние столбцы из данных:

```python
df.columns
```

```python
 df = df.drop(['BeginDate','EndDate','BeginDateDT','EndDateDT','customerID'],axis=1)
```

```python
df.head()
```

Разделим данные на целевой признак и обучающие:

```python
features = df.drop(['Target'], axis =1)
target = df['Target']
```

Выделим тестовую и обучающую выборку:

```python
features_train, features_test, target_train, target_test = train_test_split(
    features, target, test_size=0.2, random_state=1879)
```

Закодируем категориальные признаки:

```python
features_train = pd.get_dummies(features_train, drop_first=True)
features_test = pd.get_dummies(features_test, drop_first=True)
```

Проверим размер выборок:

```python
print('Размер обучающей выборки:', features_train.shape)
print('Размер валидационной выборки:', features_test.shape)
```

Маштабируем признаки:

```python
pd.options.mode.chained_assignment = None
numeric = ['MonthlyCharges','TotalCharges','TimeLive']
scaler = StandardScaler()
scaler.fit(features_train[numeric])

features_train[numeric] = scaler.transform(features_train[numeric])
features_test[numeric] = scaler.transform(features_test[numeric])
```

### Выбор модели 

Напишим функцию которая посчитает метрики и выведет кривую:

```python
def metrics(model):
    pred_proba = model.predict_proba(features_test)[:, 1]
    pred = model.predict(features_test)
    print('Accuracy: ', round(model.score(features_test, target_test), 3))
    print('ROC-AUC: ', round(roc_auc_score(target_test, pred_proba), 3))
    fpr, tpr, thresholds = roc_curve(target_test, pred_proba)
    
    sns.set()
    plt.figure(figsize=(8, 5))
    plt.plot(fpr, tpr)

    plt.plot([0, 1], [0, 1], linestyle='--')

    plt.ylim([0.0, 1.0])
    plt.xlim([0.0, 1.0])

    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')

    plt.title('ROC-кривая')
    plt.show()
    plt.figure(figsize=(8, 5))
    
    try:
        sns.barplot(y=model\
            .feature_importances_, x=features_train.columns)
        plt.xticks(rotation=90)
    except:
        pass
    return round(roc_auc_score(target_test, pred_proba), 3)   
```

Логистическая регрессия:

```python
%%time
LR = LogisticRegression(random_state=12345, class_weight = 'balanced')
LR.fit(features_train, target_train)
```

```python
RALR = metrics(LR)
```

Случайный лес:

```python
%%time
RF = RandomForestClassifier(random_state=12345)
RF.fit(features_train, target_train)
```

```python
RARF = metrics(RF)
```

```python
RARF
```

CatBoost:

```python
%%time
CAT = CatBoostClassifier(iterations=5) 
CAT.fit(features_train, target_train) 
```

```python
RACAT = metrics(CAT)
```

```python
models = {'Модель':['LogisticRegression', 'RandomForestClassifier',\
                    'CatBoostClassifier'], 
        'ROC-AUC': [RALR, RARF, RACAT],
        'Скорость обучения ms':[111, 522, 387]}
pd.DataFrame(models)
```

## Доподготовка данных и обучение модели:

Попробуем сократить лишние фичи:

Напишем функцию что бы удобно эксперементировать, 
функция будет выдовать метрики модели обученной на изменёёных данных:

```python
def search(model, x):
    features_tr_x = features_train.drop(x,axis=1) 
    features_test_x = features_test.drop(x,axis = 1)
    model = model
    model.fit(features_tr_x, target_train)
    pred_proba = model.predict_proba(features_test_x)[:, 1]
    pred = model.predict(features_test_x)
    print('Accuracy: ', round(model.score(features_test_x, target_test), 3))
    print('ROC-AUC: ', round(roc_auc_score(target_test, pred_proba), 3))
```

```python
features_train.columns
```

```python
drop = ['TotalCharges','Phone','Internet',\
        'PaymentMethod_Credit card (automatic)',\
        'gender_Male','Partner_Yes','OnlineBackup_Yes',\
       'DeviceProtection_Yes','StreamingMovies_Yes']
```

```python
search(CAT, drop)
```

Нам удалось его ещё немного улуччшить метрику.

Подготовим метрики по уже известному рецепту:

```python
features_train = features_train.drop(drop, axis=1) 
features_test = features_test.drop(drop, axis=1)
```

Проверим что получилось:

```python
features_train.columns
```

Попробуем ещё улучшить показатели модели подобрав гиперпааметры:

```python
%%time
cross_val = ShuffleSplit(n_splits=5, test_size=0.25, random_state=12345)
catboost = CatBoostClassifier(loss_function='Logloss', class_weights =[0.4,1.1])
parameters_cat = {'depth':[5,10], 'learning_rate':np.arange(0.01,0.05,0.1),
                 'iterations': range(1000,2000,200)}
catboost_grid = catboost.grid_search(parameters_cat,
            Pool(features_train, target_train),
            cv=cross_val,
            verbose=True,
            plot=False)
```

```python
%%time
final_cat = CatBoostClassifier(depth=catboost_grid["params"]["depth"], 
                        learning_rate=catboost_grid["params"]['learning_rate'],
                        iterations = catboost_grid["params"]['iterations'],
                        loss_function='Logloss',class_weights = [0.4, 1.1],
                              verbose = 10)
final_cat.fit(features_train, target_train)
```

### Тестирование модели:

```python
metrics(final_cat)
    
```
