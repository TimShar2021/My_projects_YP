## Анализ

Напишим функцию для изучения: 

```python
def stage(x, label):
    """
    принимает: x - датасет
               label - подпись
       выдаёт: функция рисует графики тренда, сезонности и остатки декомозиции.
       создает копию датасета, добовляет столюцы скользящего среднего и среднее отклонени, с окном 20
       выводит график на  полученных данных.
             
     описание: Функция визуализирует анализ временных рядов.
    """
    d = x.copy()
    d['rolling_mean'] = d['num_orders'].rolling(20).mean()
    d['rolling_std'] = d['num_orders'].rolling(20).std()
    
    decomposed = seasonal_decompose(x) 
    print('\033[1m' + label)
    print ('\033[0m')
    plt.figure(figsize=(6, 8))
    plt.subplot(311)
    decomposed.trend.plot(ax=plt.gca())
    plt.title('Trend')
    plt.subplot(312)
    decomposed.seasonal.plot(ax=plt.gca()) 
    plt.title('Seasonality')
    plt.subplot(313)
    decomposed.resid.plot(ax=plt.gca()) 
    plt.title('Residuals')
    plt.tight_layout()
    d.plot()
    plt.title('General')
    plt.tight_layout()
```

```python
stage(df, 'За всё время:')
```

```python
stage(df['2018-08':'2018-08'], 'За Август:')
```

```python
stage(df['2018-08-24':'2018-08-31'], 'Последний неделя Августа:')
```

Напишем функцию для создания новых признаков: 

```python
def make_features(df, max_lag, rolling_mean_size):
    """"
    принимает: data - датасет;
               max_lag  - максимальный размер отставания;
               rolling_mean_size - задаёт ширину окна;
               
                   
     описание: Функция создает новые признаки в данных.
    """
    
    df['hour'] = df.index.hour
    df['dayofweek'] = df.index.dayofweek
    
    for lag in range(1, max_lag + 1):
        df['lag_{}'.format(lag)] = df['num_orders'].shift(lag)
        
       
    df['rolling_mean'] = df['num_orders'].shift().rolling(rolling_mean_size-1).mean() 
```

```python
make_features(df, 1, 24)
```

## Обучение

Подготовим данные:

Разделим на признаки и целевой признак 

```python
features = df.drop('num_orders', axis = 1)
target = df['num_orders']
```

Выделим 10 % под тестовую выборку

```python
features90, features_test, target90, target_test = train_test_split(
    features, target, test_size=0.1, shuffle=False, random_state=180495)
```

Обозначим метрику RMSE:

```python
custom_score = make_scorer(mean_squared_error, greater_is_better=False)
```

Используем крос валидатор временных рядов:

```python
splits = TimeSeriesSplit(n_splits=8)
```

Обучим модель Случайного леса:

```python
model_RF = make_pipeline(StandardScaler(),RandomForestRegressor())

params_RF = {"randomforestregressor__n_estimators":[5,100],
             "randomforestregressor__max_depth":[1,10]}

grid_rf = GridSearchCV(model_RF,param_grid = params_RF, cv = splits, scoring = custom_score)
grid_rf.fit(features90,target90)
RF = grid_rf.best_estimator_ 
print('\033[1m','Обучено')
print ('\033[0m')
```

Обучим модель CatBoost:

```python
model_CAT = CatBoostRegressor(loss_function='RMSE',random_state = 180495)
params_CAT = {'learning_rate': [0.03, 0.1],
        'depth': [4, 6, 10]}

catboost_grid = model_CAT.grid_search(params_CAT,
            Pool(features90,target90),cv=splits,verbose=False)

CAT = CatBoostRegressor(depth=catboost_grid['params']['depth'], 
                              learning_rate=catboost_grid["params"]['learning_rate'],
                              loss_function='RMSE',verbose=False)
CAT.fit(Pool(features90,target90))

print('\033[1m','Обучено')
print ('\033[0m')
```

Обучим модель LightGBM:

```python
features_train, features_valid, target_train, target_valid = train_test_split(
    features90, target90, test_size=0.2, shuffle=False)

lgb_train = lgb.Dataset(features_train,target_train)
lgb_valid = lgb.Dataset(features_valid, target_valid, reference=lgb_train)
parameters_light = {'metric': 'rmse', 'max_depth':20,'random_state': 180495,'learning_rate':0.1}
Light = lgb.train(parameters_light,
                lgb_train,
                num_boost_round=1000,
                valid_sets=[lgb_train,lgb_valid],
                verbose_eval=100)

print('\033[1m','Обучено')
print ('\033[0m')
```

Константная модель:

```python
pred_median = np.ones(target_test.shape) * target90.median()
```

Выводы:

Модели обучены, можно приступать к тестированию

## Тестирование

Напишим функцию для тестирования:

```python
def rmse(model, label):
    """
    принимает: model - обученую модель;
               label - подпись;
       выдаёт: RMSE;
             
     описание: функция печатает RMSE
    """
    print('\033[1m',label)
    print ('\033[0m')
    predictions = model.predict(features_test)
    print('RMSE: {:.4f}'.format(mean_squared_error(target_test,predictions)**0.5))
    return predictions
```

Применим функцию к моделям:

```python
pRF = rmse(RF, 'Случайный лес:')
```

```python
pCB = rmse(CAT, 'CatBoost:')
```

```python
pLG = rmse(Light, 'LightGBM:')
```

```python
print('\033[1m','Констнтная модель:')
print ('\033[0m')
print('RMSE: {:.4f}'.format(mean_squared_error(target_test,pred_median)**0.5))
```

Соберем предсказания и факт в таблицу :

```python
data = {'Случайный лес' : pRF,
        'CatBoost' : pCB,
        'LightGBM' : pLG,
        'Факт' : target_test}
frame = pd.DataFrame(data)
frame.head()
```

```python
frame['2018-08-30':'2018-08-31'].plot(figsize = (16, 8))
plt.title('Сравнение предсказаний разных моделей и фактических данных')
plt.ylabel('Количество заказов такси')
```
