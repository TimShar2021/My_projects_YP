## Предобработка данных

Посмотрим первые пять строк солбца высоты отолка  с пропущенныи значениями:

```python
df[df['ceiling_height'].isnull()].head()
```

По данным не видно что они как то переклекаютя с другими столбцами совпадают пропуски только со столбцом обозначающи апартаменты а там больше всего пропусков. Думаю можно заменить медианныи заначениями. Ка видно из таблицы df.describe() 50 % квантель имеет значение 2,65 солвпадает с логическим представлением о высоте потолков в новостройках это обычно 2,5 - 2,7 м на него и заменим.

Заменим все пропуски медианными значениями и проверим:

```python
ceiling_height_median = df['ceiling_height'].median()
ceiling_height_median
df['ceiling_height'] = df['ceiling_height'].fillna(ceiling_height_median)
df.isna().sum()
```

```python
df[df['floors_total'].isnull()].head()
```

Пропусков немного можно их удалить:

```python
df = df.dropna(subset=['floors_total'])
df.isna().sum()
```

Посмотрим пропуски в данных о жилой площади:

```python
df[df['living_area'].isnull()].head()
```

Можно было бы просто заменить на мединнное значение, но тогда получатся странные данные с 3 конатными квртирами в 30 квадратных метров, попробуем использовать коэфицент.

Создадим столбец коэфицента жилой площади от общей так как значения по общей площади у нас есть:

```python
df['coef'] = df['living_area'] / df['total_area']
```

Заменим пропуски медианными значениями в зависимости от количества комнат:

```python
df['coef'].fillna(df.groupby('rooms')['coef'].transform('median'), inplace=True)
```

Наконец заполним пропуски в жилой площади площади умножив общую площадь на коэфицент:

```python
df.loc[df['living_area'].isna(), 'living_area'] = df.loc[df['living_area'].isna(), 
                                                         'total_area'] *  df.loc[df['living_area'].isna(), 'coef']
```

Проверим:

```python
df.isna().sum()
```

Заменим площадь кухонь в квартирах студиях на ноль:

```python
df.loc[(df['kitchen_area'].isna())&(df['studio'] == True)] = 0
```

```python
df.isna().sum()
```

Заменим пропуски в площади кухонь, создадим столбец соотношения алощади кухни и общей площади:

```python
df['coef_k'] = df['kitchen_area'] / df['total_area']
```

Заменим пропуски медианным значением:

```python
df['coef_k'].fillna(df.groupby('rooms')['coef'].transform('median'), inplace=True)
```

Умножим обшую площадь на коэфицент и заменим площадь кухни:

```python
df.loc[df['kitchen_area'].isna(), 'kitchen_area'] = df.loc[df['kitchen_area'].isna(), 
                                                         'total_area'] *  df.loc[df['kitchen_area'].isna(), 'coef_k']
```

Проверим:

```python
df.isna().sum()
```

Заменим значение  балконов  на ноль, так по всей видиоти у данного жилья балконов нет:

```python
df['balcony'] = df['balcony'].fillna(0)
```

Проверим какие есть значения в столбце апартаментов:

```python
df['is_apartment'].value_counts()
```

Заменим все пропуски на False так как по логике большая часть жилья не является апартаментами:

```python
df['is_apartment'] = df['is_apartment'].fillna(False)
```

Заменим типы данных в столбцах last_price, floors_total, is_apartment, studio - данные в этих столбцах удобней изучать в целых числах.А для апартаментов и студий значениями инт проще вывести среднее.

## Выведем новые параметры

Добавим в таблицу цену квадратного метра:

```python
df['price_m2'] = (df['last_price']/df['total_area']).dropna().astype('int64')
df = df.dropna(subset = ['price_m2'])
df['price_m2'] = df['price_m2'].astype('int64')
```

Приведем в порядок даты:

```python
df['first_day_exposition'] = pd.to_datetime(df['first_day_exposition'] , format ='%Y-%m-%dT%H:%M:%S')
```

Поработаем с этажами, добавим столбец с типами — первый, последний, другой, для этого напишим функцию которая будет сравнивать столбцы floor  и total_floor а так же находить первые этажи:

```python
df['floor_type'] = pd.Series() #Создадим столбец
df.head() #Проверим
```

```python
def floor_t(df):
    floor = df['floor']
    floor_total = df['floors_total']
    if floor == floor_total:
        df['floor_type'] = 'последний'
        return df['floor_type']   
    if floor == 1:
        df['floor_type'] = 'первый'
        return df['floor_type']
    else:
        df['floor_type'] = 'другой'
        return df['floor_type']
    
df['floor_type']  = df.apply(floor_t, axis=1)     
```
